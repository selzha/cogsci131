## Problem 3a

```python
def summation(n):
    result = 0
    for i in range(1, n+1):
        result += 1/(i + 1)
    return result

#test
summation(10)
```

## Problem 3b

```python
summation_function = [summation(j) for j in np.arange(0, 101)]
plt.plot(np.arange(0, 101), summation_function)
plt.show()
```

## Problem 6a

```python
np.random.seed(1)
x_6 = sorted(np.random.normal(0, 1, 1001))

array = []
for i in range(0, 1000):
    difference = abs(x_6[i+1] - x_6[i])
    array.append(difference)

final_array = array

fig, ax = plt.subplots()
ax.hist(final_array, bins=50, density=True)
plt.xlim(0, 0.3)
plt.ylim(0, 100)
plt.show()
```
